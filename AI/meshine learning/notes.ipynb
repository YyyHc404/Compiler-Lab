#  1.基本术语
###  1.1
- D={**x**<sub>1</sub>,**x**<sub>2</sub>... **x** <sub>m</sub>}   代表m个实例的数据集 
- 每个示例有d个属性也称为样本示例x的维度 
- **x**<sub>i</sub>={x<sub>i1</sub>,...,x<sub>id</sub>}
- 从数据集中学的模型的过程称为"学习"或"训练"
- 拥有了标记的示例称为"样例" (**x**<sub>i</sub>,y<sub>i</sub>)  y<sub>i</sub> $\in\gamma \gamma是所有标记的集合$
- 预测的是离散的值，称之为分类
- 预测的是连续的值，称之为回归
- 两个类别的任务，一个成为正类，一个称为反类
- 预测任务是希望通过对训练集进行学信，建立一个从输入空间 X -> Y 的映射
- 聚类，将输入分为若干组，每个组称为一个簇，存在一些潜在的规律划分
- 根据训练数据是否有标记信息，学习任务可以分为监督学习和无监督学习
- 学的模型适用于新样本的能力称为"泛化"能力
- 假设样本空间全体样本服从一个未知"分布" D,获得的每个样本都是独立的从分布上采集获得的，即"独立分布"
- 记住训练样本，就是所谓的 机械学习


### 1.2 假设空间
- 归纳:从特殊到一般化的泛化过程，数学中的公理推导
- 演绎: 从一般化到特殊的特化过程，数学中的推理推导
从样例中学习学习显然是一个归纳过程
广义的归纳表示为样例中学习，狭义的表示是从中学到概念，称为"概念学习"
- 目标是在假设空间中找出一个与训练集匹配的假设
- 任意值用 * 表示，命题不成立用空集表示



### 1.3 归纳偏好
- 机器学习算法在学习过程中对某种类型假设的偏好，称为归纳偏好
#####  奥卡姆剃刀: 
- 有多个假设与观察一致，选择更简单的哪个。

针对问题选择合适的学习算法

#####  __数学相关概念__
- 指示函数: 用来描述某个特定条件或性质是否成立的情况，通常取值是0或1
- NFL定理: 无偏见的不可能定理;在特定条件下，任何优化算法在平均情况下都没有优势。换句话说，没有一个优化算法能够在所有问题上都表现最好，即不存在一种“免费午餐”。


# 2.模型评估与选择
- 错误率e = $\frac{a}{m}$   a: 样本分类错误   m: 样本总数 
- 精度 = 1 - e 
- 误差 = 真实值 - 实际输出 
- 训练误差：在训练集上
- 泛化误差: 在新样本上
- 过拟合: 在训练集上过度概括(只有...)
- 欠拟合: 与过拟合相对

### 2.1模型选择
#### 2.1.1评估方法
通过实验测试对泛化误差进行评估。\

- 测试集:从样本空间中独立同分布采样而得，测试样本应尽可能不在训练集中出现
- 对样本集适当处理，产生训练集和测试集 
常见的划分方法:
1. 留出法: 将数据集D划分为两个互斥的集合30%的作为测试集,剩下的作为训练集
    - 分层采样:保留类别比例的采样;要确保划分的类别比例相似
    - 若干次随机划分: 按不同比例划分多次训练，取平均值作为评估结果
#####  __数学相关概念__

NP是计算复杂性理论中的一个重要概念，表示“非确定性多项式时间”（Non-deterministic Polynomial time）。NP是一类问题的集合，其中的问题可以在多项式时间内验证一个解的正确性，如果一个问题同时属于NP类并且可以在多项式时间内找到一个解，那么这个问题被认为是在P类（多项式时间可解）中。
